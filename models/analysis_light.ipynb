## Problem description
This example illustrates the use of TuLiP interface to stormpy to analyze the probability that system modelled by a Markov chain satisfies a given specification.



We begin by importing the packages and modules that we will need.

```python
import os
from tulip.interfaces import stormpy as stormpy_int
```

Set up paths to all the models

```python
model_path = os.path.join("/home/ubuntu/eeci/c1-probabilistic", "models")
light_path = os.path.join(model_path, "light.pm")
```

Build models from prism files

```python
light = stormpy_int.to_tulip_transys(light_path)
```

Specify property: "The light is eventually green"

```python
formula = 'P=? [ F ("green") ]'
```

Run model checking

```python
out_model_path = os.path.join(model_path, "out_light_model.nm")
result = stormpy_int.model_checking(light, formula, out_model_path)
```

Examine result

```python
for state in light.states:
    print("  State {}, with labels {}, Pr = {}".format(state, light.states[state]["ap"], result[state]))
```
